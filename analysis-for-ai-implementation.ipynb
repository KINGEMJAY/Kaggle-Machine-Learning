```python
import matplotlib.pyplot as plt
```

Data for the bar chart

```python
companies = ['ABC Inc', 'XYZ Corp.', 'PQR ltd.', 'DEF Co.', 'MNO Corp.']
productivity_improvement = [10, 12, 8, 15, 9]
```

Create the bar chart

```python
plt.bar(companies, productivity_improvement)
plt.xlabel('Company')
plt.ylabel('Productivity Improvement (%)')
plt.title('AI Implementation Data: Productivity Improvement by Company')
plt.show()
```
