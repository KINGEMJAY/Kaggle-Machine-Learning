Schematic of a Coastal Aquifer

```
import matplotlib.pyplot as plt

# Set up figure
fig, ax = plt.subplots(figsize=(10, 6))

# Draw aquifer layers
ax.fill_between([0, 10], 0, 3, color='tan', label='Unsaturated Zone')
ax.fill_between([0, 10], 3, 6, color='lightblue', label='Freshwater Zone')
ax.fill_between([0, 10], 6, 7, color='skyblue', hatch='//', label='Mixing Zone')
ax.fill_between([0, 10], 7, 10, color='royalblue', label='Saltwater Zone')

# Add sea and land
ax.fill_between([10, 15], 0, 8, color='navy')  # Ocean
ax.fill_between([0, 10], 0, -2, color='darkgreen')  # Land

# Well and pumping
ax.plot([5], [4], marker='o', markersize=10, color='black')  # Well
ax.arrow(5, 4, 0, -1, width=0.1, color='black')  # Pumping

# Labels and arrows
ax.text(5, 5, 'Well', ha='center')
ax.annotate('Groundwater Pumping', xy=(5, 3), xytext=(5, 1),
            arrowprops=dict(arrowstyle='->', color='black'))
ax.text(12, 4, 'Ocean', ha='center')

# Freshwater-Saltwater Interface
ax.plot([0, 10], [6, 6], 'r--', label='FSI')
ax.text(5, 6.5, 'FSI', va='bottom', ha='center', color='red')

# Tidal fluctuations
ax.annotate('Tidal Fluctuations', xy=(10, 5.5), xytext=(12, 5.5),
            arrowprops=dict(arrowstyle='<->', color='black'))

# Titles and labels
ax.set_title('Schematic of a Coastal Aquifer')
ax.set_xlabel('Distance from Shore (km)')
ax.set_ylabel('Depth (m)')
ax.legend()

# Grid
ax.grid(axis='y', linestyle='--')

plt.show()
```

Research on Groundwater Salinity in African Coastal Areas

```
import matplotlib.pyplot as plt

# Data for the figure
studies = ["Srivastava et al. (2019)", "Li et al. (2020)"]
methods = ["Field Surveys", "Laboratory Analysis", "Numerical Modelling"]
factors = ["Anthropogenic Activities", "Geological Processes", "Climatic Conditions"]

# Create the figure and subplots
fig, axs = plt.subplots(2, 1, figsize=(8, 6))
fig.suptitle("Research on Groundwater Salinity in African Coastal Areas")

# First Subplot: Past Studies and Methods
axs[0].barh(studies, [2.5, 2])  # Assuming each study used all 3 methods
axs[0].set_xlabel("Number of Studies")
axs[0].set_ylabel("Key Studies")
axs[0].set_title("Studies and Research Methods")

# Second Subplot: Factors Contributing to Salinity
axs[1].pie( [1, 1, 1], labels=factors, autopct='%1.1f%%', startangle=140)
axs[1].set_title("Factors Influencing Salinity")
axs[1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Adjust spacing and display the plot
plt.tight_layout()
plt.show()
```

Key Coastal Regions for Climate Change Impact Study

```
import matplotlib.pyplot as plt

# Data for the figure
# (Replace these with actual data if available)
regions = ['North Africa', 'West Africa', 'East Africa', 'Southern Africa']
vulnerability_score = [8.5, 9.2, 6.8, 7.5]
biodiversity_index = [0.75, 0.82, 0.65, 0.70]
population_density = [1500, 2200, 800, 1200]

# Create the figure and axes
fig, ax1 = plt.subplots()

# Bar chart for vulnerability score (primary axis)
color = 'tab:red'
ax1.set_xlabel('African Coastal Regions')
ax1.set_ylabel('Vulnerability Score', color=color)
ax1.bar(regions, vulnerability_score, color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Secondary axis for biodiversity and population density
ax2 = ax1.twinx()
ax2.set_ylabel('Biodiversity Index / Population Density', color='tab:blue')

# Line plot for biodiversity index (secondary axis)
ax2.plot(regions, biodiversity_index, color='tab:blue', marker='o', linestyle='--')

# Scatter plot for population density (secondary axis)
ax2.scatter(regions, population_density, color='tab:green', marker='x')

# Legend
lines = [plt.Line2D([0], [0], color='tab:blue', linestyle='--', marker='o'),
         plt.Line2D([0], [0], color='tab:green', marker='x')]
ax2.legend(lines, ['Biodiversity Index', 'Population Density'])

# Additional formatting
plt.title('Key African Coastal Regions for Climate Change Impact Study')
fig.tight_layout()

# Show the plot
plt.show()
```
