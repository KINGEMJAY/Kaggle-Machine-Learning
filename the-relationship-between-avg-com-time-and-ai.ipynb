```python
import matplotlib.pyplot as plt
```

Data for line graph

```python
average_completion_time = [150, 180, 200, 170, 190]
ai_adoption_level = ['High', 'Medium', 'Low']
ai_adoption_mapping = {'High': 'r','Medium':'g','Low':'b'}
```

Plotting line graph

```python
plt.figure(figsize=(8, 6))
for i, level in enumerate(ai_adoption_level): x = [j for j, value in enumerate(average_completion_time)]
y = [value for j, value in enumerate(average_completion_time)]
plt.plot(x, y, marker='o', linestyle='-', color=ai_adoption_mapping[level], label=level)
plt.xticks(range(len(average_completion_time)), [str(i) for i in average_completion_time])
plt.xlabel('Project ID')
plt.ylabel('Average Completion Time (Days)')
plt.title('Average Completion Time vs AI Adoption Level')
plt.legend()
plt.show()

```
