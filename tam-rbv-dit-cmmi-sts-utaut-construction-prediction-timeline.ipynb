```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

load dataset

```python
data = pd.DataFrame({'Project ID': [1, 2, 3, 4, 5],
                    'Actual Duration (months)': [12, 18, 24, 14, 10],
                    'Predicted Duration (months)': [11, 16, 25, 15, 11]
                    })
```

Technology Acceptance Model (TAM)

```python
data['PU'] = np.random.randint(1, 6, size=len(data))
data['PEOU'] = np.random.randint(1, 6, size=len(data))
data['Actual Usage'] = np.random.randint(1, 6, size=len(data))
```

Resource-Based View (RBV)

```python
data['Resource Availability'] = np.random.randint(1, 6, size=len(data))
```

Diffusion of Innovation Theory

```python
data['Relative Advantage'] = np.random.randint(1, 6, size=len(data))
data['Adoption'] = np.random.randint(0, 2, size=len(data))
```

Capability Maturity Model Integration

```python
data['Process Maturity'] = np.random.randint(1, 6, size=len(data))
```

```python
data['Social Factors'] = np.random.randint(1, 6, size=len(data))
```

Unified theory of acceptance and se of technology

```python
data['PE'] = np.random.randint(1, 6, size=len(data))
data['EE'] = np.random.randint(1, 6, size=len(data))
print(data)
```

visualize the relationsips

```python
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.scatter(data['PU'], data['Actual Usage'])
plt.xlabel('Perceived Usefulness')
plt.ylabel('Actual Usage')
```

```python
plt.subplot(2, 2, 2)
plt.scatter(data['PEOU'], data['Actual Usage'])
plt.xlabel('Perceived Ease of Use')
plt.ylabel('Actual Usage')           
```

```python
plt.subplot(2, 2, 3)
plt.scatter(data['Resource Availability'], data['Actual Duration (months)'])
plt.xlabel('Resource Availiability')
plt.ylabel('Actual Duration (months)')
```

```python
plt.subplot(2, 2, 4)
plt.scatter(data['Relative Advantage'], data['Adoption'])
plt.xlabel('Relative Advantage')
plt.ylabel('Adoption')
plt.tight_layout()
plt.show()
```
