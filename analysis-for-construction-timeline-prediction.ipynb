```python
import matplotlib.pyplot as plt
```

Dataset

```python
project_ids = [1, 2, 3, 4, 5]
actual_duration = [12, 18, 24, 14, 10]
predicted_duration = [11, 16, 25, 15, 11]
```

Bar Chart

```python
plt.figure(figsize=(8, 6))
plt.bar(project_ids, actual_duration, width=0.4,label='Actual Duration')
plt.bar(project_ids, predicted_duration, width=0.4, label='Predicted Duration')
plt.xlabel('Project ID')
plt.ylabel('Duration (months)')
plt.title('Construction Timeline Prediction')
plt.legend()
plt.show()
```
