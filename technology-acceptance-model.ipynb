```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

Load the dataset

```python
data = pd.DataFrame({ 'Project ID': [1, 2, 3, 4, 5],
                    'Labour Cost ($)': [100000, 150000, 200000, 120000, 180000],
                     'Material Cost ($)': [50000, 80000, 100000, 60000, 90000], 
                      'Equipment Cost ($)': [30000, 40000, 50000, 35000, 450000]
                    })
```

Calculate Total Cost

```python
data['Total Cost ($)'] = data['Labour Cost ($)'] + data['Material Cost ($)'] + data['Equipment Cost ($)']
```

Calculate perceived usefulness (PU) - average total cost

```python
pu = data['Total Cost ($)'].mean()
```

Calculate perceived ease of use (PEOU) - average labour cost

```python
peou = data['Labour Cost ($)'].mean()
```

Fit linear regression model

```python
X = np.column_stack((data['Total Cost ($)'], data['Labour Cost ($)']))
X = sm.add_constant(X)
y = np.array([pu, peou])
```

Print

```python
print('Perceived Usefulness (PU):',round(pu, 2))
```
