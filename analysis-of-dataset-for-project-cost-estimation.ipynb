```python
import matplotlib.pyplot as plt
```

Dataset

```python
project_id = [1, 2, 3, 4, 5]
actual_cost = [500000, 750000, 1000000, 650000, 900000]
predicted_cost = [510000, 720000, 980000, 670000, 910000]
```

Plotting the line graph

```python
plt.plot(project_id, actual_cost, marker='o', label='Actual Cost')
plt.plot(project_id, predicted_cost, marker='o', label='Predicted Cost')
```

Labeling the axes and title

```python
plt.xlabel('Project ID')
plt.ylabel('Cost ($)')
plt.title('Project Cost Estimation')
plt.legend()
```

Displaying the graph

```python
plt.show()
```
