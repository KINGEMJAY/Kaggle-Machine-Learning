```python
import pandas as pd
```

Load the dataset

```python
data = pd.DataFrame({'Project ID': [1, 2, 3,4, 5],
                    'Labour Cost ($)': [100000, 150000, 200000, 120000, 180000],
                    'Material Cost ($)': [50000, 80000, 100000, 60000, 90000],
                    'Equipment Cost ($)': [30000, 40000, 50000, 35000, 45000],
                    })
```

Calculate total cost

```python
data['Total Cost ($)'] = data['Labour Cost ($)'] + data['Material Cost ($)'] + data['Equipment Cost ($)']
```

Calculate resource utilization ratios

```python
data['Labour Ratio'] = data['Labour Cost ($)'] / data['Total Cost ($)']
data['Material Ration'] = data['Material Cost ($)'] / data['Total Cost ($)']
data ['Equipment Ratio'] = data['Equipment Cost ($)']  / data['Total Cost ($)']
```

Calculate average resource utilization ratios

```python
avg_labour_ratio = data['Labour Ratio'].mean()
avg_equipment_ratio = data['Equipment Ratio'].mean()
```

Print results

```python
print('Average Labour Ratio:', round(avg_labour_ratio, 2))
print('Average Equipment Ratio:', round(avg_equipment_ratio, 2))
```
