```python
import matplotlib.pyplot as plt

```

Load the dataset

```python
project_id = [1, 2, 3, 4, 5]
ai_adoption = ['High', 'Low', 'Medium', 'High', 'Low']
labour_cost_savings = [50000, 20000, 30000, 70000, 15000]
material_cost_savings = [30000, 15000, 20000, 40000, 10000]
ai_implementation_cost = [80000, 40000, 60000, 100000, 30000]
net_cost_savings = [0, -500, -10000, 10000, -5000]
                    
```

Bar chart

```python
plt.bar(project_id, net_cost_savings, color='blue', alpha=0.7)
plt.xlabel('Project ID')
plt.ylabel('Net Cost Savings ($)')
plt.title('Cost-Benefit Analysis')
plt.xticks(project_id)
plt.show()
```

Mathematical equation

```python
optimum_productivity = sum(labour_cost_savings) + sum(material_cost_savings) - sum(ai_implementation_cost)
```

```python
print(f"The optimum productivity is: {optimum_productivity}")
```
